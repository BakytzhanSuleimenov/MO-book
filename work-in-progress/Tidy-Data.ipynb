## Multi-dimensional Data

```python
import pandas as pd

data = {"Warehouse A": {"Customer 1": 254, "Customer 2": 173, "Customer 3": 330},
        "Warehouse B": {"Customer 1": 340, "Customer 2": 128, "Customer 3": 220},
        "Warehouse C": {"Customer 1": 430, "Customer 2": 250, "Customer 3": 225},
        "Warehouse D": {"Customer 1": 135, "Customer 2": 180, "Customer 3": 375}}

df = pd.DataFrame.from_dict(data).T
display(df)
```

```python
df_tidy = pd.melt(df.reset_index(), id_vars="index", value_vars=df.columns)
df_tidy.to_csv("tidy.csv", index=False)

display(df_tidy)
```
